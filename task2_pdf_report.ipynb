```python
from fpdf import FPDF
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("sales_data.csv")
summary = df.groupby("Product")["Sales"].sum().reset_index()
```

```python
class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 16)
        self.set_text_color(0, 102, 204)  # Blue
        self.cell(0, 10, '--- Sales Summary Report ---', ln=True, align='C')

        self.ln(10)
        self.set_draw_color(100, 100, 100)
        self.line(10, 20, 200, 20)

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 10)
        self.set_text_color(128)
        self.cell(0, 10, f'Generated by Python | {datetime.now().strftime("%d-%b-%Y")}', align='C')

pdf = PDF()
pdf.add_page()
pdf.set_font('Arial', 'B', 12)
pdf.set_text_color(0)
pdf.ln(5)

# Table Header
pdf.set_fill_color(220, 220, 220)
pdf.cell(60, 10, "Product", 1, 0, 'C', 1)
pdf.cell(60, 10, "Total Sales (USD)", 1, 1, 'C', 1)

# Table Rows
pdf.set_font('Arial', '', 12)
for i, row in summary.iterrows():
    pdf.cell(60, 10, str(row["Product"]), 1, 0, 'C')
    pdf.cell(60, 10, f"${row['Sales']}", 1, 1, 'C')

# Save File
pdf.output("custom_sales_report.pdf")
```
